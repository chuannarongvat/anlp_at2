```python
from requests_html import HTMLSession
import pandas as pd
from urllib.parse import urljoin
```

```python
def web_scarping(url, df, title_element, link_element, text_element):
    session = HTMLSession()
    url = url
    df = df

    response = session.get(url)
    soup = BeautifulSoup(response.html.html, "html.parser")
    news_titles = soup.find_all(title_element)
    news_links = soup.find_all(link_element)

    for i in range(len(news_titles)):
        title = news_titles[i].find('span', {'data-editable': 'headline'}).text.strip()
        news_url = urljoin(url, news_links[i]['href'])

        response = requests.get(news_url)
        soup = BeautifulSoup(response.content, 'html.parser')

        article_text = ''
        article = soup.find(text_element).find_all('p')[1:]
        article_text = "\n".join([p.get_text() for p in article])
        article_text = article_text.split('\n', 1)[1]

        df = df.append({'title': title, 'url': news_url, 'text': article_text}, ignore_index=True)
```

```python
session = HTMLSession()
url = "https://edition.cnn.com/politics"
cnn_df = pd.DataFrame(columns=['title', 'url', 'text'])

response = session.get(url)

soup = BeautifulSoup(response.html.html, "html.parser")
news_titles = soup.find_all('div', {'class': 'container__headline container_list-headlines__headline'})
news_links = soup.find_all('a', {'class': 'container__link container_list-headlines__link'})
scraped_urls = []

for i in range(len(news_titles)):
    try:
        title = news_titles[i].find('span', {'data-editable': 'headline'}).text.strip()
        url = urljoin(url, news_links[i]['href'])

        while url in scraped_urls:
            i += 1
            url = urljoin(url, news_links[i]['href'])

        scraped_urls.append(url)
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        article_text = ''
        article = soup.find('div', class_="article__content-container").find_all('p')[1:]
        article_text = "\n".join([p.get_text() for p in article])
        article_text = article_text.split('\n', 1)[1]

        
        cnn_df = cnn_df.append({'title': title, 'url': url, 'text': article_text}, ignore_index=True)
    
    except:
        continue
```

### abc news

```python
session = HTMLSession()
url = 'https://abcnews.com.co/'
abc_news_df = pd.DataFrame(columns=['title', 'description', 'url', 'article_text'])

for page in range(1, 6):
    page_url = f'{url}page/{page}/'
    r = session.get(page_url)
    articles = r.html.find('article')
    for item in articles:
        try:
            newsitem = item.find('h2', first=True)
            if not newsitem:
                newsitem = item.find('h4', first=True)

            title = newsitem.text
            link = newsitem.absolute_links
            desc = item.find('.elementor-post__excerpt p', first=True).text

            article_url = list(link)[0]
            article_r = session.get(article_url)
            article_text = article_r.html.find('.elementor-widget-theme-post-content p')
            article_text = '\n'.join([p.text for p in article_text])

            abc_news_df = abc_news_df.append({'title': title, 'description':desc, 'url': article_url, 'article_text': article_text}, ignore_index=True)
        except:
            pass
```

### infowars 

```python
url = 'https://www.infowars.com/'
infowars_news_df = pd.DataFrame(columns=['title', 'url', 'text', 'subject'])

# 2 -> Featured
# 3 -> Policitcs
# 10 -> Science and Tech
# 11 -> Economy
# 20 -> Good News
# 31 -> Editors Choice

for category in [2, 3, 10, 11, 20, 31]:
    if category == 3:
        subject = 'politics'
    else:
        subject = 'News'

    category_url = f'{url}category/{category}/'
    r = session.get(category_url)

    articles = r.html.find('a.css-1xjmleq')

    for item in articles:
        try:
            newsitem = item.find('h1', first=True)

            if newsitem:
                title = newsitem.text
                link = list(item.absolute_links)[0]

                response = requests.get(link)
                soup = BeautifulSoup(response.content, "html.parser")

                article = soup.find("div", {"class": "css-ce9o4r"}).find_all("p")
                article_text = "\n".join([p.get_text() for p in article])

                infowars_news_df = infowars_news_df.append({'title': title, 'url': link, 'text': article_text, 'subject': subject}, ignore_index=True)

        except:
            pass
```

### empire news

```python
url = "https://empirenews.net/category/politics/"
subject = 'politics'
empire_news_df = pd.DataFrame(columns=['title', 'url', 'text', 'subject'])

for page in range(1, 10):
    page_url = f'{url}page/{page}/'
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, "html.parser")
    news_titles = soup.find_all("h1", {'class': 'entry-title'})

    for title in news_titles:
        news_link = title.find('a')['href']
        title = title.text.strip()

        response = requests.get(news_link)
        soup = BeautifulSoup(response.content, "html.parser")
        article_text = ""
        article = soup.find("div", class_="entry-content").find_all("p")[1:]
        article_text = "\n".join([p.get_text() for p in article])
        article_text = article_text.split('\n', 1)[1]

        empire_news_df = empire_news_df.append({'title': title,
                                                'url': news_link,
                                                'text': article_text,
                                                'subject': subject}, ignore_index=True)
```
